# sql2_descriptive

```python
submissions_group_concat = pd.read_csv('edu - sql - 15 days of learning sql - sql2_descriptive.csv')
submissions_group_concat
```

```python
to_test = submissions_group_concat.values[0][-2]

def series_date_diff(i):
    vals = i.split(',') 
    vals_pd = [pd.to_datetime(j) 
               for j in vals]
    vals_pd_diff = [(vals_pd[k+1]-vals_pd[k]).days 
                    for k in range(len(vals_pd)-1 )]
    return vals_pd_diff


series_date_diff(to_test)
```

```python
diffs = []
vals_pd_diff_if_no_one_day_lst = []
for i in submissions_group_concat['Unnamed: 3']:
    vals_pd_diff = series_date_diff(i)
    diffs.append(vals_pd_diff)
    
submissions_group_concat['diffs']=diffs 

# submissions_group_concat['stat1'] = submissions_group_concat['diffs']!=submissions_group_concat['count(distinct submission_date)']
# submissions_group_concat[submissions_group_concat['stat1']==True]

submissions_group_concat['stat1'] = submissions_group_concat['Unnamed: 3'].astype(str).str.contains('^2016-03-01')

def slice_till_vale(lst):
    first_idx_where_not = 0
    for idx,item in enumerate(lst):
        if item!=1 and idx!=0:
            first_idx_where_not=idx-1
            break
        else:
            continue
    return first_idx_where_not
    
    
slice_to = []
for i in submissions_group_concat['diffs']:
    try:
        val = slice_till_vale(i)
        slice_to.append(val)
    except:
        slice_to.append(None)
        continue

submissions_group_concat['slice_to']=slice_to
submissions_group_concat[submissions_group_concat['slice_to']!=0]
```

```python
pd.options.display.max_colwidth=500
```

```python
def slice_to_idx(lst_as_str,idx):
    lst = lst_as_str.split(',')
    idx = idx+2
    return lst[:idx]

sliced = []
for idx_slicer,lst_as_str in zip(submissions_group_concat['slice_to'],submissions_group_concat['Unnamed: 3']):
    if idx_slicer>0:
        res = slice_to_idx(lst_as_str,idx_slicer)
        sliced.append(res)
    else:
        sliced.append(None)
        
        
submissions_group_concat['sliced']=sliced
submissions_group_concat[ (submissions_group_concat['slice_to']!=0)
                         &(submissions_group_concat['stat1']==True)]#['sliced']
```

```python
diffs = []
vals_pd_diff_if_no_one_day_lst = []
for i in submissions_group_concat['Unnamed: 3']:
    vals = i.split(',') 
    vals_pd = [pd.to_datetime(j) for j in vals]
    vals_pd_diff = [str(vals_pd[k+1]-vals_pd[k]) for k in range(len(vals_pd)-1 )]
    vals_pd_diff_concat = ','.join(vals_pd_diff)
    #vals_pd_diff_if_no_one_day = ','.join([d for d in vals_pd_diff if str(d)!='1 days 00:00:00'])
    diffs.append(vals_pd_diff)
    #vals_pd_diff_if_no_one_day_lst.append(vals_pd_diff)
    
submissions_group_concat['diffs']=diffs  
#submissions_group_concat['vals_pd_diff_if_no_one_day_lst']=vals_pd_diff_if_no_one_day_lst  
```
